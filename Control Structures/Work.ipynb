```python
###
# Checking whether a car exceeded the speed limit
#
speed_limit = 140
car_speed = int( input('Enter car speed (km/h): ') )

if car_speed > speed_limit:
    print(f'Your speed is {car_speed}km/h')
    print('Warning: speed limit exceeded!!')
else:
    print(f'Your speed is {car_speed}km/h')
    print("You're driving safely, thank you!")
```

```python
###
# Credit card payment 
#
account_balance = 500
payment_list = [140, 499, 500, 501, 720]

for total_payment in payment_list:
    print (f'Trying to pay: {total_payment}')
    if total_payment <= account_balance:
        print('Payment completed')
    else:
        print('No funds')
```
